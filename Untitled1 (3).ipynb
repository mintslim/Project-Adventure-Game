```python
import random

print("Welcome to the Treasure Adventure game!")
print("You are lost in a forest and need to find the treasure to ensure survival!")
print("You can move North, South, East, or West.")

treasure_location = (random.randint(1, 10), random.randint(1, 10))
player_location = (random.randint(1, 10), random.randint(1, 10))
moves_left = 20
health = 3

obstacles = [(random.randint(1, 10), random.randint(1, 10)) for _ in range(5)]

while moves_left > 0:
    print("You are at location", player_location)
    print("You have", moves_left, "moves left.")
    print("Your health is", health)
    
    if player_location == treasure_location:
        print("Congratulations! You found the treasure!")
        break
    
    if player_location in obstacles:
        print("Oh no! You hit a dangerous obstacle!")
        health -= 1
        if health == 0:
            print("Sorry, you died. Game over.")
            break
        
    direction = input("Which direction would you like to move? (N/S/E/W) ")
    
    if direction == "N":
        player_location = (player_location[0], player_location[1] + 1)
    elif direction == "S":
        player_location = (player_location[0], player_location[1] - 1)
    elif direction == "E":
        player_location = (player_location[0] + 1, player_location[1])
    elif direction == "W":
        player_location = (player_location[0] - 1, player_location[1])
    else:
        print("Wrong direction. Please enter N, S, E, or W.")
        continue
    
    moves_left -= 1
    
    if moves_left == 0:
        print("Sorry, you're out of moves. Game over.")
        break
    
    distance_from_treasure = abs(player_location[0] - treasure_location[0]) + abs(player_location[1] - treasure_location[1])
    print("You are", distance_from_treasure, "moves away from the treasure.")
    
print("Thanks for playing!")





```
